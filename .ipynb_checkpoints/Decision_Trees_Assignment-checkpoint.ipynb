| Num_posts        | Number of total posts that the user has ever posted   |
| Num_following    | Number of following                                   |

| Average_hashtag_count   | Average number of hashtags used in a post |
| Promotional_keywords | Average use of promotional keywords in hashtag, i.e. regrann, contest, repost, giveaway, mention, share, give away, quiz |
| Followers_keywords | Average use of followers hunter keywords in hashtag, i.e. follow, like, folback, follback, f4f|
| Cosine_similarity  | Average cosine similarity of between all pair of two posts a user has |
| Post_interval      | Average interval between posts (in hours) |

# Random Forests

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# write your code here ^_^
Random_FC = RandomForestClassifier(n_estimators=50)
```

```python
Random_FC.fit(X_train,y_train)
```

```python
RFC_predictors = Random_FC.predict(X_test)
```

```python
print(classification_report(y_test,RFC_predictors))
```

```python
confusion_matrix(y_test,RFC_predictors)
```

```python
plot_confusion_matrix(Random_FC,X_test,y_test)
```

```python
print('Accuracy score :',accuracy_score(y_test,RFC_predictors))
```
